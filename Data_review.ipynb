```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Selfie-dataset/selfie_dataset.csv')
df.head(5)
```

```python
print('Black mean popularity ',df[df.black == 1].popularity_score.mean())
print('White mean popularity ',df[df.white == 1].popularity_score.mean())
print('Asian mean popularity ',df[df.asian == 1].popularity_score.mean())
```

```python
print('Round face mean popularity ',df[df.round_face == 1].popularity_score.mean())
print('Oval face mean popularity ',df[df.oval_face == 1].popularity_score.mean())
print('Heart face mean popularity ',df[df.heart_face == 1].popularity_score.mean())
```

```python
print('Male mean popularity ',df[df.is_female == 0].popularity_score.mean())
print('Female mean popularity ',df[df.is_female == 1].popularity_score.mean())
```

```python
print('baby mean popularity ',df[df.baby == 1].popularity_score.mean())
print('child mean popularity ',df[df.child == 1].popularity_score.mean())
print('teenager mean popularity ',df[df.teenager == 1].popularity_score.mean())
print('youth mean popularity ',df[df.youth == 1].popularity_score.mean())
print('middle age mean popularity ',df[df.middle_age == 1].popularity_score.mean())
print('senior mean popularity ',df[df.senior == 1].popularity_score.mean())
```

```python
print('Female with lipstick mean popularity ',df[(df.wearing_lipstick == 1) & (df.is_female == 1)].popularity_score.mean())
print('Female with no lipstick mean popularity ',df[(df.wearing_lipstick == 0) & (df.is_female == 1)].popularity_score.mean())
```

```python
print("Female")
print('black hair mean popularity ',df[(df.black_hair == 1) & (df.is_female == 1)].popularity_score.mean())
print('braid hair mean popularity ',df[(df.braid_hair == 1) & (df.is_female == 1)].popularity_score.mean())
print('brown hair mean popularity ',df[(df.brown_hair == 1) & (df.is_female == 1)].popularity_score.mean())
print('straight hair mean popularity ',df[(df.straight_hair == 1) & (df.is_female == 1)].popularity_score.mean())
print('curly hair mean popularity ',df[(df.curly_hair == 1) & (df.is_female == 1)].popularity_score.mean())
print('blond hair mean popularity ',df[(df.blond_hair == 1) & (df.is_female == 1)].popularity_score.mean())
print('red hair mean popularity ',df[(df.red_hair == 1) & (df.is_female == 1)].popularity_score.mean())
print("Male")
print('brown hair mean popularity ',df[(df.brown_hair == 1) & (df.is_female == 0)].popularity_score.mean())
print('black hair mean popularity ',df[(df.black_hair == 1) & (df.is_female == 0)].popularity_score.mean())
print('straight hair mean popularity ',df[(df.straight_hair == 1) & (df.is_female == 0)].popularity_score.mean())
print('curly hair mean popularity ',df[(df.curly_hair == 1) & (df.is_female == 0)].popularity_score.mean())
print('blond hair mean popularity ',df[(df.blond_hair == 1) & (df.is_female == 0)].popularity_score.mean())
print('braid hair mean popularity ',df[(df.braid_hair == 1) & (df.is_female == 0)].popularity_score.mean())
print('red hair mean popularity ',df[(df.red_hair == 1) & (df.is_female == 0)].popularity_score.mean())
```

```python
print('duckface mean popularity ',df[(df.duck_face  == 1)].popularity_score.mean())
print('no duckface mean popularity ',df[(df.duck_face  == 0)].popularity_score.mean())
```

```python
plt.hist(df.popularity_score, color = 'blue', edgecolor = 'black',
         bins = 100)

# seaborn histogram
sns.distplot(df.popularity_score, hist=True, kde=False, 
             bins=100, color = 'blue',
             hist_kws={'edgecolor':'black'})
# Add labels
plt.title('Distribution density')
plt.xlabel('Popularity score')
plt.ylabel('number of photos')
```
